```python
import pandas as pd
```

```python
# importing data for year 2019

df_19 = pd.read_csv('./201906-citibike-tripdata.csv')
df_19
```

```python
# importing data for the year 2020
df_20= pd.read_csv('./202006-citibike-tripdata.csv')
df_20
```

```python
#combining the two imported data into one single df
combined_df= pd.concat([df_19,df_20],axis=0)
print (combined_df)
```

```python
# Define a mapping dictionary
gender_mapping = {0: 'Unknown', 1: 'Male', 2: 'Female'}

# Apply the mapping to the 'gender' column
combined_df['gender'] = combined_df['gender'].map(gender_mapping)

# Display the updated DataFrame
print(combined_df.head())
```

```python
#transferring the clean data to new csv 
combined_df.to_csv('citibike.csv',index=False)
```
